```python
# PART 4
# Mostrar columna 'tags' y crear columna 'tag_conteo' que enumere las etiquetas de cada video
dfv['tags']
dfv['tag_conteo']=dfv['tags'].apply(lambda x: len(x.split('|')) if (x!="[none]") else 0)
dfv['tag_conteo']
dfv
```

```python
# PART 5
# Crear columna con nombre "Likes", que contendrá los booleanos True y False.
# True si el número de likes supera el de dislikes

dfv['dif_likes']=dfv['likes']-dfv['dislikes']
dfv['dif_likes']
dfv.head()

def thumbs_up(valor):
    if  valor>0:
        return True
    else:
        return False
dfv['bool_likes']=dfv['dif_likes'].apply(thumbs_up)
dfv
```

```python
# PART 6
# Convertir el archivo .json en df
import json
with open('US_category_id.json', "r") as f:data=json.loads(f.read())
data
```

```python
arch2=pd.json_normalize(data, record_path=['items'])
arch2
```

```python
arch2.rename(columns={'id':'category_id', 'snippet.title':'category_title'}, inplace=True)
```

```python
id_titles=arch2[['category_id','category_title']]
id_titles
```

```python
# PART 7
# Combinar dataframe original con el nuevo y mantener solamente las columnas:
# 'title','channel_title','category_title','views','likes','dislikes','comment_count','description','tag_count','thumbs_up'
```

```python
arch2['category_id']=arch2['category_id'].astype(int)
```

```python
dfv=dfv.merge(arch2,how='inner',left_on='category_id',right_on='category_id')
dfv=dfv[['title','channel_title','category_title','views','likes','dislikes','comment_count','description', 'tag_conteo', 'bool_likes']]
```

```python
dfv
```

```python
# PART 8
# Mostrar el top 20 de videos por views con la siguiente información: 
# número dentro del top (rank), vistas (views), nombre del video. Subrayar cada resultado
```

```python
df_top20=dfv.sort_values(by=['views'], ascending=False).head(20)
df_top20
df_top20['rank']=np.arange(len(df_top20))
df_top20['rank']=df_top20['rank']+1
```

```python
# PART 9
# Mostrar un gráfico de densidad del conteo de tags en color verde con título
```

```python
# PART 10
# Mostrar un histograma de likes vs dislikes en color azul con título
```

```python
# PART 11
# Mostrar una gráfica de barras con el número de veces que se presenta cada categoria
dfv['category_title'].value_counts().plot.bar(color='red', figsize=(10,5)).set_title("Count of category titles");
```
