```python
# 1. Obtener el apellido, el cargo y el salario de todos los empleados cuyo trabajo es el de "Programmer" o "Shipping Clerk"
# y cuyo salario no sea igual a 4,500, 10,000 o 15,000 dólares
sql="""select em.last_name as 'Apellido', j.job_title as 'Cargo', em.salary as 'Salario'
from employees em
inner join jobs j on j.job_id=em.job_id
where j.job_title in ('Programmer', 'Shipping Clerk')
and (salary!= '4500' or salary!= '10000' or salary!= '15000')
"""
pd.read_sql_query(sql,conn)
```

```python
# 4. Obtener el salario máximo de un empleado que trabaja como 'programmer'
sql="""select max(e.salary) as 'Salario_Máximo' from employees e
inner join jobs j on j.job_id=e.job_id
where j.job_title in ('Programmer')"""
pd.read_sql_query(sql,conn)
```

```python
# 5. Obtener el salario medio y el número de empleados que trabajan en el departamento 90.
sql="""select avg(salary) as 'SalarioMedio', count(employee_id) as '#Empleados' from employees
where department_id like 90
"""
pd.read_sql_query(sql,conn)
```

```python
# 6. Obtener el salario más alto, el más bajo, la suma y la media de todos los salarios.
sql="""select max(salary) as 'SalarioMáximo', min(salary) as 'SalarioMínimo', 
sum(salary) as 'SumaSalario', avg(salary) as 'SalarioMedio' from employees
"""
pd.read_sql_query(sql,conn)
```

```python
# 7. Obtener el número de empleados con el mismo puesto.
sql="""
select j.job_title as 'Puesto', count(e.employee_id) as '#Empleados'
from jobs j
inner join employees e on e.job_id=j.job_id
group by j.job_title
order by count(e.employee_id) DESC"""
pd.read_sql_query(sql,conn)
```

```python
# 1. Obtener nombre, apellido y salario de los empleados cuyo salario es igual al salario mínimo de su rol.
sql="""
select e.first_name, e.last_name, e.salary, j.job_title from employees e
inner join jobs j on j.job_id=e.job_id
where e.salary=j.min_salary """
pd.read_sql_query(sql,conn)
```

```python
# 2. Obtener nombre, apellido y salario de los empleados que ganan más que el salario medio y que trabajan en cualquier área de IT
sql="""
select e.first_name, e.last_name, e.salary, d.depart_name from employees e
inner join departments d on d.department_id=e.department_id
where d.depart_name like 'IT'
and e.salary > (select avg(salary) from employees)
"""
pd.read_sql_query(sql,conn)
```

```python
# 3. Obtener nombre, apellido y salario de los empleados que ganan más que el Sr. Bell.
df= pd.read_sql_query("select first_name, last_name, salary from employees where salary> (select salary from employees where last_name='Bell')",conn)
df
```

```python
# 4. Obtener nombre y apellido de los empleados que no son supervisors
sql="""
select e.first_name as 'Nombre', e.last_name as 'Apellido' from employees e
inner join jobs j on j.job_id=e.job_id
where j.job_title not like '%Supervisor%'
"""
pd.read_sql_query(sql,conn)
```

```python
# Nadie cuenta con un job_title que contenga 'supervisors' pero si lo que se quiere consultar son aquellos que no son gerentes:
# 4. Obtener nombre y apellido de los empleados que no son managers
sql="""
select e.first_name as 'Nombre', e.last_name as 'Apellido' from employees e
inner join jobs j on j.job_id=e.job_id
where j.job_title not like '%Manager%'
"""
pd.read_sql_query(sql,conn)
```

```python
# 5. Obtener el employee ID, nombre, apellido y salario de todos los empleados cuyo salario es superior a la media de sus departamentos
sql="""
select e.employee_id as 'ID Empleado', e.first_name as 'Nombre', e.last_name as 'Apellido', e.salary as 'Salario' from employees e
left join (select avg(salary) as 'aveg_salary', department_id from employees medio group by department_id) AS medio
on medio.department_id = e.department_id
where e.salary > aveg_salary
"""
pd.read_sql_query(sql,conn)
```

```python
# 6. Obtener el 5° salario máximo de la tabla de empleados.
df=pd.read_sql_query("select * from employees", conn);
df=df.sort_values(by=['salary'])
df.salary.nlargest(5).iloc[-1]
```

```python
# 7. Obtener las direcciones (location_id, street_address, city, state_province, country_name) de todos los departamentos.
sql="""
select d.depart_name, l.location_id, l.location_id, l.street_address, l.city, l.state_province, c. country_name from locations l
inner join departments d on d.location_id=l.location_id
inner join countries c on c.country_id=l.country_id
order by d.depart_name ASC
"""
pd.read_sql_query(sql, conn)
```

```python
# 8. Obtener el “employee ID”, nombre y apellido de cada empleado junto con el nombre, apellido y “manager ID” de su “manager”.
sql="""
select e.employee_id as 'ID Empleado', e.first_name as 'Nombre', e.last_name as 'Apellido', man.first_name as 'Nombre Manager', man.last_name as 'Apellido Manager', e.manager_id as 'ID Manager'
from employees e
left join (select first_name, last_name, employee_id from employees) as man
on man.employee_id=e.manager_id 
"""
pd.read_sql_query(sql, conn) #aqui el join se hace con diferente campo (employee_id=manager_id)
```

```python
# 9. Obtener el job_title y el salario medio de los empleados.
sql="""
select j.job_title as 'Puesto', avg(e.salary) as 'Salario Medio' from employees e
inner join jobs j on j.job_id=e.job_id
group by j.job_title
"""
pd.read_sql_query(sql, conn)
```
